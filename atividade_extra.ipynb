```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
caminho_arquivo = r"C:\Users\gabriel_e_machado\Documents\Programação de aplicativos\correcao_bugs.xlsx"

df_bugs = pd.read_excel(caminho_arquivo)
```

```python
df_bugs
```

```python
x = df_bugs[["Localização do Bug","Sistema Operacional","Versão do Software","Número de Ocorrências","Tempo Médio de Resolução","Complexidade do Bug"]]
y = df_bugs[["Classificação de Bug"]]
```

```python
mapeamento_loc = {
    "Frontend":0,
    "Backend":1,
    "Banco de Dados":2,
    "Integrações Externas":3,
    "Configuração do Ambiente":4,
    "Segurança":5,
    "Performance":6,
    "UI/UX":7,
    "Compatibilidade de Navegador":8,
    "Mobile":9
}

mapeamento_sys = {
    "Windows":0,
    "Linux":1
}

mapeamento_ocor = {
    "Única ocorrência":0,
    "Ocasional":1,
    "Muitas":2,
    "Recorrente":3,
    "Frequente":4
}

mapeamento_temp = {
    "Menos de 1 dia":0,
    "1-3 dias":1,
    "3-7 dias":2
}

mapeamento_complex = {
    "Simples":0,
    "Moderada":1,
    "Complexa":2
}
```

```python
x["Localização do Bug"] = x["Localização do Bug"].map(mapeamento_loc)
x["Sistema Operacional"] = x["Sistema Operacional"].map(mapeamento_sys)
x["Número de Ocorrências"] = x["Número de Ocorrências"].map(mapeamento_ocor)
x["Tempo Médio de Resolução"] = x["Tempo Médio de Resolução"].map(mapeamento_temp)
x["Complexidade do Bug"] = x["Complexidade do Bug"].map(mapeamento_complex)
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=37)

modelo = DecisionTreeClassifier()
modelo.fit(x_train,y_train)

previsoes = modelo.predict(x_test)
```
